```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd

PATH='C:\Program Files (x86)\chromedriver.exe'
driver=webdriver.Chrome(PATH)
rankings_url='https://www.atptour.com/en/rankings/singles'



try:
    driver.get(rankings_url)
    mega_table=WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "mega-table"))
        )
    
    ranking_data=dict()
    mega_table_headers=mega_table.find_element_by_tag_name('thead')
    
    
    for header in mega_table_headers.find_elements_by_class_name('sorting-label'):
        ranking_data[header.text]=[]
    
    #print(ranking_data)
    
    ranking_data.pop('Move')
    ranking_data.pop('Country')
    ranking_data.pop('Points Dropping')
    ranking_data.pop('Next Best')
    
    ranking_data['profile_url']=[]
    
    
    mega_table_body=mega_table.find_element_by_tag_name('tbody')
    
    for player_data in mega_table_body.find_elements_by_tag_name('tr'):
        
        ranking_data['Ranking'].append(player_data.find_element_by_class_name('rank-cell').text)
        ranking_data['Player'].append(player_data.find_element_by_class_name('player-cell').text)
        ranking_data['profile_url'].append(player_data.find_element_by_class_name('player-cell').find_element_by_tag_name('a').get_attribute('href'))
        ranking_data['Age'].append(player_data.find_element_by_class_name('age-cell').text)
        ranking_data['Tourn Played'].append(player_data.find_element_by_class_name('tourn-cell').text)
        ranking_data['Points'].append(player_data.find_element_by_class_name('points-cell').text)
        
    
    driver.quit()
    

except Exception as e:
    print(e)
    
    driver.quit()



    
    
    
```

```python
import time
from tqdm import tqdm
from selenium.webdriver.chrome.options import Options


rankings_history=[]
stats=[]

for player,profile_url in tqdm(zip(ranking_data['Player'],ranking_data['profile_url'])):

    PATH='C:\Program Files (x86)\chromedriver.exe'
    driver=webdriver.Chrome(PATH)
    driver.get(profile_url)
    
    
    profile_tabs=WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "profileTabs"))
        )
    
    
    stats.append(profile_tabs.find_element_by_link_text("Player Stats").get_attribute('href'))
    rankings_history.append(profile_tabs.find_element_by_link_text("Rankings History").get_attribute('href'))
    
    
    
    
    driver.quit()
    
    
```

```python
count=1
for url in rankings_history:
    PATH='C:\Program Files (x86)\chromedriver.exe'
    driver=webdriver.Chrome(PATH)
    driver.get(url)
    
    mega_table=WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "mega-table"))
        )
    
    ranking_data=dict()
    mega_table_header_sec=mega_table.find_element_by_tag_name('thead')
    mega_table_header_row=mega_table_header_sec.find_element_by_tag_name('tr')
    headers=mega_table_header_row.find_elements_by_tag_name('th')
    
    for header in headers:
        ranking_data[header.text]=[]
        
    
    mega_table_body=mega_table.find_element_by_tag_name('tbody')
    
    
    for row in mega_table_body.find_elements_by_tag_name('tr'):
        for key,data in zip(ranking_data.keys(),row.find_elements_by_tag_name('td')):
            ranking_data[key].append(data.text)
            
    
    ranking_df=pd.DataFrame.from_dict(ranking_data)
    
    
    ranking_df.to_csv(f'Ranking_hist_{count}.csv',index=False)
    
    driver.quit()
    count+=1
    
```

```python
count=1

stats1=dict()
stats2=dict()


for url in stats:
    #print(stats1)
    #print(stats2)
    PATH='C:\Program Files (x86)\chromedriver.exe'
    driver=webdriver.Chrome(PATH)
    driver.get(url)
    
    mega_tables=WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "mega-table"))
        )     
    
    stats1_body=mega_tables[0].find_element_by_tag_name('tbody')
    
    
    for row in stats1_body.find_elements_by_tag_name('tr'):
        data=row.find_elements_by_tag_name('td')
        #print(data[0].text in stats1)
        if data[0].text in stats1:
            stats1[data[0].text].append(data[1].text)
        else:
            stats1[data[0].text]=[data[1].text]
        
    
    print('\n\n')
    stats2_body=mega_tables[1].find_element_by_tag_name('tbody')
    
    for row in stats2_body.find_elements_by_tag_name('tr'):
        data=row.find_elements_by_tag_name('td')
        #print(data[0].text in stats2)
        
        if data[0].text in stats2:
            stats2[data[0].text].append(data[1].text)
        else:
            stats2[data[0].text]=[data[1].text]
    
    print('\n\n')
    driver.quit()
    count+=1
    
```

```python
stat1_df=pd.DataFrame.from_dict(stats1)
```

```python
stat1_df.to_csv('Serve_Stats.csv')
```

```python
stat2_df=pd.DataFrame.from_dict(stats2)
stat2_df.to_csv('Return_Stats.csv')
```
